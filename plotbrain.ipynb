To plot the centroids of the Schaefer 100 for report diagram

```python
import numpy as np
import matplotlib.pyplot as plt
from nilearn import plotting, datasets

# Load the Schaefer 100 parcellation
atlas = datasets.fetch_atlas_schaefer_2018(n_rois=100)
atlas_filename = atlas['maps']
labels = atlas['labels']

# Load the coordinates of the centers of the parcels
coords = plotting.find_parcellation_cut_coords(labels_img=atlas_filename)

# Plot the glass brain
display = plotting.plot_glass_brain(None, display_mode='lyrz')

# Add nodes (dots)
for coord in coords:
    display.add_markers([coord], marker_color='r', marker_size=50)

plt.show()
```
